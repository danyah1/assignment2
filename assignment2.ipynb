## Dataset Summary

The Titanic dataset contains information about passengers such as age, gender, class, and ticket fare.

### Machine Learning Problem
The goal of this dataset is to predict whether a passenger survived or not.

### Type of Problem
This is a supervised classification problem because the target variable Survived has two classes:
- 0 = Did not survive  
- 1 = Survived

This dataset can be used to train a model that learns patterns from passenger features.
